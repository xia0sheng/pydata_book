```python
import pandas as pd
import os
encoding = 'latin1'

upath = os.path.expanduser('ch02/movielens/users.dat')
rpath = os.path.expanduser('ch02/movielens/ratings.dat')
mpath = os.path.expanduser('ch02/movielens/movies.dat')

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table(upath, sep='::', header=None, names=unames, encoding=encoding, engine='python')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table(rpath, sep='::', header=None, names=rnames, encoding=encoding, engine='python')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table(mpath, sep='::', header=None, names=mnames, encoding=encoding, engine='python')
```

```python
users[:5]
```

```python
ratings[:5]
```

```python
movies[:5]
```

```python
ratings
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data
```

```python
data.loc[0]
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings[:5]
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
top_female_ratings[:10]
```

```python
rating_std_by_title = data.groupby('title')['rating'].std()
rating_std_by_title = rating_std_by_title.loc[active_titles]
rating_std_by_title = rating_std_by_title.sort_values(ascending=False)
rating_std_by_title[:10]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

%matplotlib inline

npath = os.path.expanduser('ch02/names/yob1880.txt')
names1880 = pd.read_csv(npath, names=['name', 'sex', 'births'])

names1880
```

```python
names1880.groupby('sex').births.sum()
```

```python
years = range(1880, 2011)

pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = os.path.expanduser('ch02/names/yob%d.txt' % year)
    frame = pd.read_csv(path, names=columns)
    
    frame['year'] = year
    pieces.append(frame)
    
names = pd.concat(pieces, ignore_index=True)
```

```python
names
```

```python
total_births = names.pivot_table('births', index='year', columns='sex', aggfunc=sum)
```

```python
total_births
```

```python
total_births.plot(title='Total births by sex and year')
```

```python
def add_prop(group):
    births = group.births
    
    group['prop'] = births / births.sum()
    return group
names = names.groupby(['year', 'sex']).apply(add_prop)
```

```python
names
```

```python
np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)
```

```python
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]
grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
```

```python
top1000
```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
```

```python
total_births = top1000.pivot_table('births', index='year', columns='name', aggfunc=sum)
total_births
```

```python
subset = total_births[['John', 'Harry', 'Mary', 'Marilyn']]
subset.plot(subplots=True, figsize=(12, 10), grid=False, title='Number of births per year')
```

```python
table = top1000.pivot_table('prop', index='year', columns='sex', aggfunc=sum)
table.plot(title='Sum of top1000.prop by year and sex', yticks=np.linspace(0, 1.2, 13), xticks=range(1880, 2020, 10))
```
